```python
import openai
import pandas as pd
from datasets import load_dataset
```

```python

# Initialize the OpenAI API key (replace YOUR_API_KEY with your actual key)
openai.api_key = "YOUR_API_KEY"

# Load the HumanEval dataset
ds = load_dataset("openai/openai_humaneval")

# Helper function to generate code from GPT-4
def generate_code_from_gpt(prompt):
    try:
        response = openai.Completion.create(
            engine="gpt-4",
            prompt=prompt,
            max_tokens=500,  # Adjust the max tokens based on expected response length
            n=1,
            stop=None,
            temperature=0.5
        )
        return response.choices[0].text.strip()  # Extract the generated code
    except Exception as e:
        return str(e)

# Initialize a list to store task data and LLM responses
data = []

# Iterate through each task in the dataset
for i, task in enumerate(ds['test']):
    task_id = task['task_id']
    prompt = task['prompt']
    
    # Generate code from GPT-4 based on the task prompt
    gpt_code = generate_code_from_gpt(prompt)
    print(f"Generated code for task {task_id}:")
    print(gpt_code)
    
    # Store task ID, prompt, and generated code in the data list
    data.append({
        'task_id': task_id,
        'prompt': prompt,
        'gpt_code': gpt_code
    })

# Convert the data list to a pandas DataFrame and save it as a CSV
df = pd.DataFrame(data)
df.to_csv('llm_responses.csv', index=False)

print("LLM responses have been saved to llm_responses.csv")
```
